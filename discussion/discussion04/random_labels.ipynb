# ROC curves with random labels

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, roc_auc_score

plt.style.use("ggplot")
```

```python
np.random.seed(42)
n_samples = 1000
d = 10
p = 0.8
X = np.random.randn(n_samples, d)
y = np.random.binomial(1, p, n_samples)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
```

```python
lin_model = LogisticRegression()
lin_model.fit(X_train, y_train)
fpr, tpr, _ = roc_curve(y_test, lin_model.predict_proba(X_test)[:,1])
```

```python
plt.figure(figsize = (8,8))
plt.plot(fpr, tpr)
#plt.scatter(results_df["FPR"], results_df["TPR"])
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.title(f"ROC curve (FPR vs TPR) with p = {p}")
plt.show()
```
